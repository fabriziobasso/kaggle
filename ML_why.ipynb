# **MACHINE LEARNING vs TRADITIONAL STATISTICS**

## **1.0 Brave New World**
---

**Traditional statistical methods** typically make assumptions about the underlying relationship between variables before fitting a model to the data. These assumptions may be based on theoretical principles or prior knowledge about the data. For example, a linear regression model assumes that the relationship between two variables is linear, which means that the change in one variable is proportional to the change in the other variable.

**Machine learning methods**, on the other hand, do not make assumptions about the underlying relationship between variables. Instead, they learn the relationship from the data itself. This means that machine learning models can be used to model relationships that are not linear or that are not well-understood by traditional statistical methods.

Traditional statistics and machine learning differ in their approach to defining and estimating the "betas" of a regression model.

**Traditional Statistics**

In traditional statistics, the betas () represent the coefficients of a linear regression model, which is a statistical model that predicts a continuous target variable (y) based on one or more independent variables (x). The betas are estimated using a technique called least squares estimation, which minimizes the sum of squared errors between the predicted values of y and the actual values of y.

Traditional statistical methods assume that the relationship between the independent variables and the target variable is linear and that the errors are independent and identically distributed (IID). These assumptions allow for the derivation of statistical properties of the beta estimates, such as their standard errors and confidence intervals.

**Machine Learning**

Machine learning approaches to regression can involve a wider range of models, including non-linear models and models with complex interactions between variables. The estimation of betas in machine learning often relies on optimization algorithms that seek to minimize a loss function, such as mean squared error or mean absolute error.

Machine learning methods do not necessarily require assumptions about the linearity of the relationship or the distribution of errors. This flexibility allows machine learning to model complex relationships that may not be well-captured by traditional linear regression models. However, it also makes it more challenging to interpret the betas and derive their statistical properties.

## **NEW PROBLEMS WE CAN SOLVE**
---



Classic types of problems for Machine Learning can be broadly categorized into three main areas:

1. **Supervised Learning**: In supervised learning, the machine learning model is trained on a dataset of labeled examples, where each example consists of an input data point and its corresponding target value. The goal of supervised learning is to learn a mapping from input data points to target values, so that the model can make accurate predictions for new, unseen data.
2. **Unsupervised Learning**: In unsupervised learning, the machine learning model is trained on a dataset of unlabeled data, where the data points do not have any associated target values. The goal of unsupervised learning is to uncover hidden patterns or structure in the data. This can involve tasks such as clustering, dimensionality reduction, and anomaly detection.
3. **Reinforcement Learning**: In reinforcement learning, the machine learning model learns through trial and error, interacting with an environment and receiving rewards for actions that lead to desired outcomes. The goal of reinforcement learning is to learn a policy that maximizes the cumulative reward over time.

Here are some specific examples of classic machine learning problems:

* **Classification**: Classifying data points into predefined categories. For example, classifying emails as spam or not spam, or classifying images of handwritten digits.

* **Regression**: Predicting a continuous numerical value based on input data points. For example, predicting housing prices based on features such as square footage, number of bedrooms, and location.

* **Clustering**: Grouping data points into clusters based on their similarity. For example, clustering customers based on their purchase history to identify different customer segments.

* **Dimensionality Reduction**: Reducing the number of features in a dataset while preserving as much of the information as possible. This can be useful for tasks such as data visualization and machine learning modeling.

* **Anomaly Detection**: Identifying data points that deviate significantly from the normal patterns. For example, detecting fraudulent transactions in financial data, or detecting network intrusions in cybersecurity.

* **Recommendation Systems**: Recommending products, movies, music, or other items to users based on their past preferences and behavior. For example, recommending products to online shoppers based on their browsing history, or recommending movies to viewers based on their past ratings.

These classic machine learning problems have a wide range of applications across various domains, including healthcare, finance, marketing, manufacturing, and more. As machine learning techniques continue to evolve, we can expect to see even more innovative and impactful applications in the future.

```
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.colors import ListedColormap

from sklearn.datasets import make_circles, make_classification, make_moons
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn.gaussian_process import GaussianProcessClassifier
from sklearn.gaussian_process.kernels import RBF
from sklearn.inspection import DecisionBoundaryDisplay
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.neural_network import MLPClassifier
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier

names = [
    "Nearest Neighbors",
    "Linear SVM",
    "RBF SVM",
    "Gaussian Process",
    "Decision Tree",
    "Random Forest",
    "Neural Net",
    "AdaBoost",
    "Naive Bayes",
    "QDA",
]

classifiers = [
    KNeighborsClassifier(3),
    SVC(kernel="linear", C=0.025, random_state=42),
    SVC(gamma=2, C=1, random_state=42),
    GaussianProcessClassifier(1.0 * RBF(1.0), random_state=42),
    DecisionTreeClassifier(max_depth=5, random_state=42),
    RandomForestClassifier(
        max_depth=5, n_estimators=20, max_features=1, random_state=42
    ),
    MLPClassifier(alpha=1, max_iter=1000, random_state=42),
    AdaBoostClassifier(random_state=42),
    GaussianNB(),
    QuadraticDiscriminantAnalysis(),
]

X, y = make_classification(
    n_features=2, n_redundant=0, n_informative=2, random_state=1, n_clusters_per_class=1
)
rng = np.random.RandomState(2)
X += 2 * rng.uniform(size=X.shape)
linearly_separable = (X, y)

datasets = [
    make_moons(noise=0.3, random_state=0),
    make_circles(noise=0.2, factor=0.5, random_state=1),
    linearly_separable,
]

figure = plt.figure(figsize=(27, 9))
i = 1
# iterate over datasets
for ds_cnt, ds in enumerate(datasets):
    # preprocess dataset, split into training and test part
    X, y = ds
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.4, random_state=42
    )

    x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
    y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

    # just plot the dataset first
    cm = plt.cm.RdBu
    cm_bright = ListedColormap(["#FF0000", "#0000FF"])
    ax = plt.subplot(len(datasets), len(classifiers) + 1, i)
    if ds_cnt == 0:
        ax.set_title("Input data")
    # Plot the training points
    ax.scatter(X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k")
    # Plot the testing points
    ax.scatter(
        X_test[:, 0], X_test[:, 1], c=y_test, cmap=cm_bright, alpha=0.6, edgecolors="k"
    )
    ax.set_xlim(x_min, x_max)
    ax.set_ylim(y_min, y_max)
    ax.set_xticks(())
    ax.set_yticks(())
    i += 1

    # iterate over classifiers
    for name, clf in zip(names, classifiers):
        ax = plt.subplot(len(datasets), len(classifiers) + 1, i)

        clf = make_pipeline(StandardScaler(), clf)
        clf.fit(X_train, y_train)
        score = clf.score(X_test, y_test)
        DecisionBoundaryDisplay.from_estimator(
            clf, X, cmap=cm, alpha=0.8, ax=ax, eps=0.5
        )

        # Plot the training points
        ax.scatter(
            X_train[:, 0], X_train[:, 1], c=y_train, cmap=cm_bright, edgecolors="k"
        )
        # Plot the testing points
        ax.scatter(
            X_test[:, 0],
            X_test[:, 1],
            c=y_test,
            cmap=cm_bright,
            edgecolors="k",
            alpha=0.6,
        )

        ax.set_xlim(x_min, x_max)
        ax.set_ylim(y_min, y_max)
        ax.set_xticks(())
        ax.set_yticks(())
        if ds_cnt == 0:
            ax.set_title(name)
        ax.text(
            x_max - 0.3,
            y_min + 0.3,
            ("%.2f" % score).lstrip("0"),
            size=15,
            horizontalalignment="right",
        )
        i += 1

plt.tight_layout()
plt.show()
```
